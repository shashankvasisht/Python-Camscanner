```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import rect
%matplotlib inline
```

```python
img = cv2.imread('test_s1.jpg')
print img.shape
```

```python
img = cv2.resize(img, (1500,880))

# create a copy of the original image
orig = img.copy()
```

```python
plt.figure(0)
plt.imshow(img)
plt.show()
```

```python
# Step 1: To Do Edge Detection.

#1.1 Change to GrayScale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#1.2 Blurring for Smoothness :- Options-> Gaussian Blur, Median Blur
gray = cv2.GaussianBlur(gray, (5,5), 0)
#1.3 Applying the Edge Detection..
edged = cv2.Canny(gray, 75, 100)
plt.figure(0)
plt.imshow(img)
plt.figure(1)
plt.imshow(gray)
plt.figure(2)
plt.imshow(edged)
plt.show()
```
